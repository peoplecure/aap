```python
import spacy
```

```python
processor = spacy.load('en')
processed_text = processor(text)
```

```python
# Show the most common words in the article.
from collections import Counter

# all tokens that arent stop words or punctuations
words = [token.text 
             for token in processed_text 
                 if token.is_punct != True]

word_freq = Counter(words)
common_words = word_freq.most_common(10)
common_words
```

```python
n = 0
for sentence in processed_text.sents:
    for token in sentence:
        print(n, token, token.pos_ == "NOUN")
        n+=1
```

```python
# Show the most common words under a part of speech. (i.e. NOUN: {'Bob':12, 'Alice':4,})
nouns = [token.text 
             for token in processed_text 
                 if token.is_stop != True 
                 and token.is_punct != True 
                 and token.pos_ == "NOUN"]

noun_freq = Counter(nouns)
common_nouns = noun_freq.most_common(10)
common_nouns
```

```python
# Find a subject/object relationship through the dependency parser in any sentence.
def pr_tree(word, level):
    if word.is_punct:
        return
    for child in word.lefts:
        pr_tree(child, level+1)
    print('\t'* level + word.text + ' - ' + word.dep_)
    for child in word.rights:
        pr_tree(child, level+1)
        
for sentence in processed_text.noun_chunks: # changed .sents to .noun_chunks to narrow parsing
    pr_tree(sentence.root, 0)
    print('-------------------------------------------')
```

```python
# Spacy tutorial method
for chunk in processed_text.noun_chunks:
    print(chunk.text, 
          chunk.root.text, 
          chunk.root.dep_)
```

```python
# Show the most common Entities and their types.
entities = [token.text + " " + token.label_
             for token in processed_text.ents     
           ]
entities_freq = Counter(entities)
common_entities = entities_freq.most_common(10)
common_entities
```

```python
# Find Entites and their dependency (hint: entity.root.head)
for entity in processed_text.ents:
    print(entity.label_, entity.root.head)
```

```python
# Find the most similar words in the article
mylist= [(token1.text, token2.text, token1.similarity(token2)) for token2 in processed_text for token1 in processed_text]

import pandas as pd

df = pd.DataFrame(mylist)
df.columns = ["Token1", "Token2", "Similarity"]
condition = (df.Similarity == 1) # excluding complete similarity (identical word comparisons)
excluded = df[~condition]
sort_top = excluded.sort_values("Similarity", ascending=False)
sort_top
```
