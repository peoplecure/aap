## 1. Create a 4 dimensional data set with 64 elements and show 2D plots of the data $x_1 \rightarrow y, x_2 \rightarrow y$, etc.

```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline

n = 64
x = np.linspace(0, 1, n) + np.random.rand(4, n)
x = np.vstack([x, np.ones(len(x.T))]).T
y = np.linspace(0, 1, n) + np.random.rand(n) - 1

plt.scatter(x.T[0], y) # First dimension
plt.scatter(x.T[1], y) # Second dimension
plt.scatter(x.T[2], y) # Third dimension
plt.scatter(x.T[3], y) # Fourth dimension
```

## 2.  Create a model to fit the data. Hint: follow the example from Lesson 3

```python
# $\beta = (X^T X)^{-1} Y^T X$

# The manual method you showed us to calculate Bet

left = np.linalg.inv(np.dot(x.T, x))
right = np.dot(y.T, x)
np.dot(left, right)
```

```python
# numpy method - it creates an error message for me, 
# but the output is correct. I am unsure what this error message is about

beta = np.linalg.lstsq(x, y)[0]
beta
```

```python
pred = np.dot(x, beta)
pred
```

## 4. Read in `mlnn/data/Credit.csv` with Pandas and create a model to predict Credit Rating (`Rating`). Use only the numeric columns in your model, but feel free to experiment which which columns you believe are better predicters of Credit Rating

```python
import pandas as pd
credit = pd.read_csv('../data/Credit.csv')
credit.head()
```

```python
X = credit[['Income', 'Limit', 'Balance', 'Cards']].values
X = np.vstack([X.T, np.ones(len(X))]).T

y = credit['Rating']

plt.scatter(X.T[0], y, c='red') # Should be income?
plt.scatter(X.T[1], y, c='orange') # Limit? 
plt.scatter(X.T[2], y, c='yellow') # Balance?
plt.scatter(X.T[3], y, c='purple') # Cards ?
plt.scatter(X.T[4], y, c='black') # Cards ?
```

```python
left = np.linalg.inv(np.dot(X.T, X))
right = np.dot(y.T, X)
np.dot(left, right)

beta = np.linalg.lstsq(X, y)[0]

pred = np.dot(X, beta)
pred

```
