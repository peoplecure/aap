# Assignment
1. change the `encoding_dim` through various values (`range(2,18,2)` and store or keep track of the best loss you can get. Plot the 8 pairs of dimensions vs loss on a scatter plot
2. using the previous assignment's model of detecting images, how does the accuracy change when you run the digit-prediction model on these 'decoded' values?
3. apply noise to *_only_* the input of the autoencoder (not the output). demonstrate that your autoencoder can strip out noise.

```python
#package import statements
from keras.callbacks import TensorBoard
from keras.layers import Input, Dense
from keras.models import Model
from keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt
import time

NAME = "autoencoder-encoding-dim-{}".format(int(time.time()))
tensorboard = TensorBoard(log_dir=r'C:\Users\cbroker1\AppData\Local\Temp\autoencoder\{}'.format(NAME))
```

# Datasets for encoding_dim values (range(2,18,2)):

Parent Dataset:

```python
(xtrain, ytrain), (xtest, ytest) = mnist.load_data()

xtrain = xtrain.astype('float32') / 255.
xtest = xtest.astype('float32') / 255.
xtrain = xtrain.reshape((len(xtrain), np.prod(xtrain.shape[1:])))
xtest = xtest.reshape((len(xtest), np.prod(xtest.shape[1:])))
xtrain.shape, xtest.shape
```

Child Datasets:

```python
#encoding_dim = 2
xtrain2 = xtrain.copy()
xtest2 = xtest.copy()

#encoding_dim = 4
xtrain4 = xtrain.copy()
xtest4 = xtest.copy()

#encoding_dim = 6
xtrain6 = xtrain.copy()
xtest6 = xtest.copy()

#encoding_dim = 8
xtrain8 = xtrain.copy()
xtest8 = xtest.copy()

#encoding_dim = 10
xtrain10 = xtrain.copy()
xtest10 = xtest.copy()

#encoding_dim = 12
xtrain12 = xtrain.copy()
xtest12 = xtest.copy()

#encoding_dim = 14
xtrain14 = xtrain.copy()
xtest14 = xtest.copy()

#encoding_dim = 16
xtrain16 = xtrain.copy()
xtest16 = xtest.copy()

#encoding_dim = 18
xtrain18 = xtrain.copy()
xtest18 = xtest.copy()
```

# Models:

Encoding_dim = 2

```python
NAME = "autoencoder-encoding-dim-2-{}".format(int(time.time()))
tensorboard = TensorBoard(log_dir=r'C:\Users\cbroker1\AppData\Local\Temp\autoencoder\{}'.format(NAME))

# this is the size of our encoded representations
encoding_dim = 2  

# this is our input placeholder
x = input_img = Input(shape=(784,))
# "encoded" is the encoded representation of the input
x = Dense(256, activation='relu')(x)
x = Dense(128, activation='relu')(x)
encoded = Dense(encoding_dim, activation='relu')(x)


# "decoded" is the lossy reconstruction of the input
x = Dense(128, activation='relu')(encoded)
x = Dense(256, activation='relu')(x)
decoded = Dense(784, activation='sigmoid')(x)

# this model maps an input to its reconstruction
autoencoder = Model(input_img, decoded)

encoder = Model(input_img, encoded)

# create a placeholder for an encoded (32-dimensional) input
encoded_input = Input(shape=(encoding_dim,))
# retrieve the last layer of the autoencoder model
dcd1 = autoencoder.layers[-1]
dcd2 = autoencoder.layers[-2]
dcd3 = autoencoder.layers[-3]

# create the decoder model
decoder = Model(encoded_input, dcd1(dcd2(dcd3(encoded_input))))

autoencoder.compile(optimizer='adadelta', loss='binary_crossentropy', metrics=['accuracy'])

autoencoder.fit(xtrain2, xtrain2,
                epochs=1000,
                batch_size=256,
                shuffle=True,
                validation_data=(xtest2, xtest2),
                callbacks=[tensorboard])

score2 = autoencoder.evaluate(xtest2, xtest2, verbose=0)
print('Test Loss:', score2[0])
print('Test Acc:', score2[1])
```
